```python
# Function to safely format ticker symbols
def get_ticker(company):
    if isinstance(company, str):  # Ensure it's a string before applying operations
        return company.strip().replace(" ", "") + ".NS"  # Assuming NSE exchange
    return None  # Return None if the value is NaN

# Apply the function safely
stats['Ticker'] = stats['COMPANY'].apply(get_ticker)

# Drop rows where ticker is None (invalid company names)
stats = stats.dropna(subset=['Ticker'])

# Fetch latest prices using the corrected tickers
latest_prices = {}

for ticker in stats['Ticker'].unique():
    try:
        stock_data = yf.download(ticker, period="1d")
        if not stock_data.empty:
            latest_prices[ticker] = stock_data['Close'].iloc[-1].item()  # Extract scalar value
    except Exception as e:
        latest_prices[ticker] = None

# Map latest prices to original dataframe
stats['Latest Price'] = stats['Ticker'].map(latest_prices)

# Ensure profit/loss calculations also return only numeric values
stats['Profit/Loss'] = ((stats['Latest Price'] - stats['PURCHASE PRICE']) * stats['QUANTITY']).astype(float)
stats['Percentage Change'] = (((stats['Latest Price'] - stats['PURCHASE PRICE']) / stats['PURCHASE PRICE']) * 100).astype(float)

# Save to a new CSV
output_file = r'B:\HELLO\Investment-Analysis\Profit_Loss_Report.csv'
stats.to_csv(output_file, index=False)

print(f" Profit/Loss report saved as: {output_file}")
```
