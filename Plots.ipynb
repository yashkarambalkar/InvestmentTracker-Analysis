```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load CSV file
def load_data(file_path):
    try:
        df = pd.read_csv(file_path)
        print("Data loaded successfully!")
        return df
    except Exception as e:
        print(f"Error loading data: {e}")
        return None
```

```python
# Plot Profit and Loss
def plot_profit_loss(df):
    if df is None:
        print("No data to plot.")
        return

    # Check for required columns
    if 'DATE' not in df.columns or 'Profit/Loss' not in df.columns:
        print("CSV must contain 'DATE' and 'Profit/Loss' columns.")
        return

    # Convert Date to datetime
    df['DATE'] = pd.to_datetime(df['DATE'])

    # Plot data
    plt.figure(figsize=(12, 6))
    plt.plot(df['DATE'], df['Profit/Loss'], label='Profit/Loss', color='blue', marker='o')

    plt.xlabel('Date')
    plt.ylabel('Profit/Loss')
    plt.title('Profit and Loss Over Time')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python


# Main Function
def main():
    file_path = "B:\HELLO\Investment-Analysis\Profit_Loss_Report.csv"
    df = load_data(file_path)
    plot_profit_loss(df)

if __name__ == "__main__":
    main()
```
